## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

## Difference Between T-test and Z-test

The t-test and the z-test are both statistical tests used to determine whether there is a significant difference between the means of two groups. The main differences between the two tests lie in the sample size, population variance, and the specific conditions under which they are used.

### T-test:

1. **Used when the sample size is small (typically \( n < 30 \) ).**
2. **The population variance is unknown.**
3. **It uses the t-distribution, which accounts for the additional uncertainty due to the smaller sample size.**

### Example Scenario for a T-test:
Suppose a teacher wants to know if a new teaching method is more effective than the traditional method. She tests the new method on a small group of 15 students and the traditional method on another small group of 15 students. Because the sample size is small and the population variance is unknown, a t-test would be appropriate to compare the means of the two groups.

### Z-test:

1. **Used when the sample size is large (typically \( n $\geq$ 30 \)).**
2. **The population variance is known or the sample size is large enough for the Central Limit Theory to apply, making the sample variance a good estimate of the population variance.**
3. **It uses the z-distribution (normal distribution).**

### Example Scenario for a Z-test:
Imagine a researcher wants to compare the average heights of adult men in two different countries. He collects data from a large sample of 1000 men from each country. Since the sample size is large and the population variance is known (or can be accurately estimated), a z-test would be appropriate to determine if there is a significant difference in the average heights between the two countries.

## Summary:
- **T-test**: Small sample size, unknown population variance.
- **Z-test**: Large sample size, known population variance or large enough sample size to estimate population variance.

### Example Calculation for Each Test:

### T-test Example:

- **Sample size (n)**: 15 students in each group
- **Sample means (\( $ \bar{X}_1 $ \) and \( $ \bar{X}_2 $ \)  )**: Mean scores of the two groups
- **Sample standard deviations (s1 and s2)**: Standard deviations of the two groups

The t-statistic is calculated as:
\[$ t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\left(\frac{s_1^2}{n_1}\right) + \left(\frac{s_2^2}{n_2}\right)}} $ \]

### Z-test Example:

- **Sample size (n)**: 1000 men in each country
- **Sample means (  \( $\bar{X}_1$\) and \($\bar{X}_2$\)  )**: Mean heights of the two samples
- **Population standard deviation (\( $ \sigma $ \))**: Known population standard deviation

The z-statistic is calculated as:
\[ z = $ \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\left(\frac{\sigma^2}{n_1}\right) + \left(\frac{\sigma^2}{n_2}\right)}}$ \]


## Q2: Differentiate between one-tailed and two-tailed tests.

### Difference Between One-tailed and Two-tailed Tests

Statistical tests can be classified into one-tailed and two-tailed tests based on the direction of the hypothesis being tested. The choice between a one-tailed and a two-tailed test depends on the research question and the specific hypothesis.

### One-tailed Test

1. **Definition**: A one-tailed test evaluates the direction of the effect. It tests whether the sample mean is either significantly greater than or significantly less than the population mean (but not both).
2. **Hypotheses**:
   - **Right-tailed test**: \( $ H_0: \mu \leq \mu_0 $ \) vs. \( $ H_1: \mu > \mu_0 $ \)
   - **Left-tailed test**: \( $ H_0: \mu \geq \mu_0 $ \) vs. \( $ H_1: \mu < \mu_0 $ \)
3. **Critical Region**: The critical region is located in one tail of the distribution (either the left tail or the right tail).

### Example Scenario for a One-tailed Test:
A pharmaceutical company claims that a new drug increases the average recovery rate from a disease. The null hypothesis \( $ H_0 $ \) states that the drug does not increase the recovery rate, while the alternative hypothesis \( $ H_1 $ \) states that the drug does increase the recovery rate. Since the interest is only in whether the drug increases the recovery rate, a one-tailed test would be appropriate.

### Two-tailed Test

1. **Definition**: A two-tailed test evaluates both directions of the effect. It tests whether the sample mean is significantly different from the population mean, regardless of the direction (greater than or less than).
2. **Hypotheses**:
   - \( $ H_0: \mu = \mu_0 $ \) vs. \( $ H_1: \mu \neq \mu_0 $ \)
3. **Critical Region**: The critical region is located in both tails of the distribution (both the left tail and the right tail).

### Example Scenario for a Two-tailed Test:
A quality control manager wants to know if a machine produces items with an average weight different from the specified standard weight of 500 grams. The null hypothesis \( $ H_0 $ \) states that the average weight is 500 grams, while the alternative hypothesis \( $ H_1 $ \) states that the average weight is not 500 grams. Since the interest is in any deviation from the standard weight, a two-tailed test would be appropriate.

### Summary:
- **One-tailed test**: Tests for an effect in one direction (greater than or less than).
- **Two-tailed test**: Tests for an effect in either direction (different from).

### Visual Representation:




````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Define the range for the x-axis
x = np.linspace(-4, 4, 1000)

# Define the normal distribution
y = norm.pdf(x, 0, 1)

# Plotting the One-tailed Test (Right-tailed)
plt.figure(figsize=(12, 6))

# One-tailed (Right-tailed)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b', label='Normal Distribution')
plt.fill_between(x, y, where=(x > 1.645), color='red', alpha=0.5, label='Critical Region')
plt.title('One-tailed Test (Right-tailed)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Two-tailed
plt.subplot(1, 2, 2)
plt.plot(x, y, 'b', label='Normal Distribution')
plt.fill_between(x, y, where=(x > 1.96) | (x < -1.96), color='red', alpha=0.5, label='Critical Region')
plt.title('Two-tailed Test')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()
````

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.


## Type 1 and Type 2 Errors in Hypothesis Testing

In hypothesis testing, two types of errors can occur: Type 1 error and Type 2 error. These errors are related to the incorrect rejection or acceptance of the null hypothesis.

## Type 1 Error (False Positive)

1. **Definition**: A Type 1 error occurs when the null hypothesis (\( $ H_0 $ \)) is true, but it is incorrectly rejected. This means that a significant effect is found when there is none.
2. **Probability**: The probability of committing a Type 1 error is denoted by \( $ \alpha $ \), which is the significance level of the test (e.g., 0.05).

### Example Scenario for Type 1 Error:
Suppose a medical researcher is testing a new drug to determine if it is effective in treating a disease. The null hypothesis (\( $ H_0 $ \)) states that the drug has no effect. A Type 1 error would occur if the researcher concludes that the drug is effective when, in fact, it is not. This could lead to the approval and use of an ineffective drug.

## Type 2 Error (False Negative)

1. **Definition**: A Type 2 error occurs when the null hypothesis (\( $ H_0 $ \)) is false, but it is incorrectly accepted. This means that no significant effect is found when there is one.
2. **Probability**: The probability of committing a Type 2 error is denoted by \( $ \beta $ \). The power of a test, which is \( $ 1 - \beta $\), represents the probability of correctly rejecting the null hypothesis when it is false.

### Example Scenario for Type 2 Error:
Imagine a quality control manager is testing whether a machine produces parts that meet a specific standard. The null hypothesis (\( $ H_0 $ \)) states that the parts produced meet the standard. A Type 2 error would occur if the manager concludes that the parts meet the standard when, in reality, they do not. This could result in defective parts being used in production.

## Summary:

- **Type 1 Error (False Positive)**: Rejecting a true null hypothesis (\( $ H_0 $ \)). Probability is \( $ \alpha $ \).
- **Type 2 Error (False Negative)**: Accepting a false null hypothesis (\( $ H_0 $ \)). Probability is \( $ \beta $ \).

## Q4: Explain Bayes's theorem with an example.

## Bayes's Theorem

Bayes's theorem is a fundamental concept in probability theory and statistics, which describes the probability of an event based on prior knowledge of conditions that might be related to the event. It provides a way to update the probability of a hypothesis as more evidence or information becomes available.

The theorem is mathematically stated as:

\[$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$\]

Where:
- \($P(A|B)$\) is the posterior probability: the probability of event \($ A $\) occurring given that \($ B $\) is true.
- \($P(B|A)$\) is the likelihood: the probability of event \($ B $\) occurring given that \($ A $\) is true.
- \($ P(A) $\) is the prior probability: the initial probability of event \($ A $\) occurring.
- \($ P(B) $\) is the marginal likelihood: the total probability of event \($ B $\) occurring.

## Example Scenario:

### Medical Test Example

Suppose a certain disease affects 1% of the population. There is a medical test for this disease, which is 99% accurate. This means:
- If a person has the disease, the test will correctly identify it 99% of the time (true positive rate).
- If a person does not have the disease, the test will correctly return a negative result 99% of the time (true negative rate).

Let's define the events:
- \($ D $\): The event that a person has the disease.
- \($ T $\): The event that the test result is positive.

Given:
- \($ P(D) = 0.01 $\)
- \($ P(T|D) = 0.99 $\)
- \($ P(T|\neg D) = 0.01 $\), where \( $\neg D $\) represents the event of not having the disease.

We want to find the probability that a person has the disease given that they tested positive, \($ P(D|T) $\).

Using Bayes's theorem:

\[$P(D|T) = \frac{P(T|D) \cdot P(D)}{P(T)}$\]

First, we need to find \($ P(T) $\):
\[$P(T) = P(T|D) \cdot P(D) + P(T|\neg D) \cdot P(\neg D)$\]

Substitute the values:
\[$P(T) = (0.99 \cdot 0.01) + (0.01 \cdot 0.99) = 0.0099 + 0.0099 = 0.0198$\]

Now, calculate \($ P(D|T)$\):
\[$P(D|T) = \frac{0.99 \cdot 0.01}{0.0198} = \frac{0.0099}{0.0198} = 0.5$\]

So, the probability that a person has the disease given that they tested positive is 0.5 or 50%.

### Interpretation:
Despite the test being 99% accurate, the probability of having the disease given a positive test result is only 50%. This is because the disease is very rare (1% prevalence), so the number of false positives is relatively high compared to the number of true positives.

## Summary:
Bayes's theorem provides a powerful tool for updating probabilities based on new evidence. It is widely used in various fields, including medicine, finance, and machine learning, to make informed decisions under uncertainty.


## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

### Calculating a Confidence Interval

A confidence interval is a range of values that likely contains the true unknown parameter of interest, such as a population mean or proportion, based on a sample from the population and a chosen level of confidence.

#### Steps to Calculate a Confidence Interval

1. **Determine the Sample Mean (\(\bar{x}\)) and Sample Standard Deviation (\(s\)):**
   - Collect a sample of data from the population.
   - Compute the sample mean (\(\bar{x}\)) and the sample standard deviation (\(s\)).

2. **Choose the Confidence Level (CL):**
   - Decide on a confidence level, often denoted as \(1 - \alpha\), where \(\alpha\) is the significance level (typically 0.05 for 95% confidence).

3. **Find the Critical Value (z-score or t-score):**
   - Determine the critical value from the standard normal distribution (z-score) or the Student's t-distribution (t-score) based on the sample size and chosen confidence level.

4. **Calculate the Margin of Error (MOE):**
   - Use the formula:
     \[
     MOE = \text{Critical Value} \times \frac{s}{\sqrt{n}}
     \]
     where \(n\) is the sample size.

5. **Construct the Confidence Interval:**
   - Construct the confidence interval using the formula:
     \[
     \text{Confidence Interval} = \left( \bar{x} - MOE, \bar{x} + MOE \right)
     \]

#### Example:

Suppose you want to estimate the average height of students in a school with 95% confidence. You collect a sample of 50 students and find their average height to be 170 cm with a standard deviation of 5 cm.

1. **Sample Information:**
   - Sample mean (\(\bar{x}\)): 170 cm
   - Sample standard deviation (\(s\)): 5 cm
   - Sample size (\(n\)): 50

2. **Confidence Level:**
   - CL = 95%, so \(1 - \alpha = 0.95\) which implies \(\alpha = 0.05\).

3. **Critical Value:**
   - For a 95% confidence interval with \(n = 50\), the critical value from the t-distribution (since \(n < 30\)) is approximately 2.009 (you can find this from t-tables or using software).

4. **Margin of Error (MOE):**
   \[
   MOE = 2.009 \times \frac{5}{\sqrt{50}} \approx 1.42 \text{ cm}
   \]

5. **Confidence Interval:**
   \[
   \text{Confidence Interval} = (170 - 1.42, 170 + 1.42) = (168.58, 171.42)
   \]

Therefore, you can be 95% confident that the true average height of students in the school lies between 168.58 cm and 171.42 cm based on this sample data.


## Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

### Using Bayes' Theorem

Bayes' Theorem allows us to update our beliefs about the probability of an event occurring based on new evidence. It is expressed as:

\[$ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)}$ \]

where:
- \($ P(A | B)$ \) is the probability of event \( A \) occurring given that event \( B \) has occurred.
- \($ P(B | A)$ \) is the probability of event \( B \) occurring given that event \( A \) has occurred.
- \($ P(A) $\) is the prior probability of event \( A \).
- \( $P(B) $\) is the prior probability of event \( B \).

#### Sample Problem:

Suppose a factory produces two types of products: type A and type B. Historically, type A products have a defect rate of 10%, while type B products have a defect rate of 5%. If 60% of the products produced are type A and 40% are type B, what is the probability that a randomly selected defective product came from type A?

#### Solution:

Let's define the events:
- \( A \): Product is type A.
- \( B \): Product is defective.

Given:
- \( $P(A) = 0.60$ \) (Probability that a product is type A)
- \($ P(B | A) = 0.10 $\) (Probability that a type A product is defective)
- \($ P(B | \neg A) = 0.05 $\) (Probability that a type B product is defective)
- \($ P(\neg A) = 1 - P(A) = 0.40 $\) (Probability that a product is type B)

First, calculate \( P(B) \), the total probability of a product being defective:

\[$ P(B) = P(B | A) \cdot P(A) + P(B | \neg A) \cdot P(\neg A) $\]
\[ $P(B) = 0.10 \cdot 0.60 + 0.05 \cdot 0.40 $ \]
\[$ P(B) = 0.06 + 0.02 $\]
\[$ P(B) = 0.08 $\]

Now, apply Bayes' Theorem to find \( $P(A | B) $\), the probability that a defective product came from type A:

\[ $P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)}$ \]
\[ $P(A | B) = \frac{0.10 \cdot 0.60}{0.08} $\]
\[ $P(A | B) = \frac{0.06}{0.08} $\]
\[ $P(A | B) = 0.75 $\]

Therefore, the probability that a randomly selected defective product came from type A is \( $\boxed{0.75} $\) or \($ \boxed{75\%} $\).


## Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.

### Calculating the 95% Confidence Interval

To calculate the 95% confidence interval for a sample mean (\(\bar{x}\)) with known standard deviation (\(s\)), use the formula:

\[$ \text{Confidence Interval} = \left( \bar{x} - Z \cdot \frac{s}{\sqrt{n}}, \bar{x} + Z \cdot \frac{s}{\sqrt{n}} \right) $\]

where:
- \( $\bar{x} $\) is the sample mean,
- \( s \) is the sample standard deviation,
- \( n \) is the sample size,
- \( Z \) is the critical value from the standard normal distribution corresponding to the desired confidence level (for 95% confidence, \( Z \approx 1.96 \)).

#### Given Data:

- Sample mean (\($\bar{x}$\)): 50
- Sample standard deviation (\(s\)): 5
- Sample size (\(n\)): Typically not specified for this calculation (assumed to be sufficiently large)

#### Calculation:

1. **Critical Value (Z-score)**:
   - For a 95% confidence interval, \($ Z \approx 1.96 $\).

2. **Margin of Error (MOE)**:
   \[$   MOE = Z \cdot \frac{s}{\sqrt{n}}$   \]
   Since \( n \) is not explicitly given, we proceed assuming a large sample size where \($ \sqrt{n} \approx 1 $\).

   \[$   MOE = 1.96 \cdot \frac{5}{\sqrt{n}} \approx 1.96 \cdot \frac{5}{1} = 9.8$   \]

3. **Confidence Interval**:
   \[$   \text{Confidence Interval} = (50 - 9.8, 50 + 9.8) = (40.2, 59.8)$   \]

#### Interpretation:

The 95% confidence interval for the population mean (\($\mu$\)) based on this sample is from approximately 40.2 to 59.8. This means that we are 95% confident that the true population mean falls within this range. In practical terms, if we were to repeatedly sample and calculate the confidence intervals in this manner, we would expect approximately 95% of these intervals to contain the true population mean.

This interval provides a measure of the uncertainty surrounding our sample mean estimate, taking into account the variability inherent in sampling.


## Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

### Margin of Error in a Confidence Interval

The margin of error (MOE) in a confidence interval measures the amount of uncertainty or variability around an estimate, such as a sample mean or proportion. It is influenced by the sample size (\( n \)), the standard deviation of the population (if known), and the chosen level of confidence.

#### Formula for Margin of Error:

For a population mean (\($ \mu $\)) with known standard deviation (\($ \sigma $\)):

\[ $\text{MOE} = Z \cdot \frac{\sigma}{\sqrt{n}} $\]

where:
- \( Z \) is the critical value from the standard normal distribution corresponding to the desired confidence level.
- \( $\sigma $\) is the population standard deviation.
- \( n \) is the sample size.

#### How Sample Size Affects Margin of Error:

- **Inverse Relationship:** The margin of error decreases as the sample size increases, assuming all other factors remain constant. This is because larger samples provide more precise estimates of the population parameter.

#### Example Scenario:

Suppose we want to estimate the average score on a standardized test for a population of students. We take two different sample sizes and calculate the resulting margins of error:

1. **Sample Size 1:** \($ n_1 = 100 $\)
   - Given: Standard deviation of test scores, \($ \sigma = 10 $\)
   - Desired confidence level: 95% (\( $Z \approx 1.96$ \))

   Calculate MOE:
   \[$   \text{MOE}_1 = 1.96 \cdot \frac{10}{\sqrt{100}} = 1.96 \cdot 1 = 1.96$   \]

   Therefore, with a sample size of 100, the margin of error is approximately \($ \pm 1.96 $\) points.

2. **Sample Size 2:** \( $n_2 = 400$ \)
   - Given: Same standard deviation of test scores, \($ \sigma = 10 $\)
   - Desired confidence level: 95% (\( $Z \approx 1.96$ \))

   Calculate MOE:
   \[$   \text{MOE}_2 = 1.96 \cdot \frac{10}{\sqrt{400}} = 1.96 \cdot 0.5 = 0.98$   \]

   With a sample size of 400, the margin of error decreases to approximately \($ \pm 0.98 $\) points.

#### Interpretation:

In this example, increasing the sample size from 100 to 400 significantly reduces the margin of error from \($ \pm 1.96 $\) points to \($ \pm 0.98 $\) points. This means that with a larger sample size, we have greater confidence that our sample mean score is closer to the true population mean score. Larger sample sizes improve the precision of our estimates and reduce the uncertainty around the estimated parameter.


## Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.


### Calculating the Z-score

The z-score measures how many standard deviations a data point is away from the mean of a population. It is calculated using the formula:

\[ $ z = \frac{x - \mu}{\sigma} $\]

where:
- \( x \) is the value of the data point,
- \($ \mu $\) is the population mean,
- \($ \sigma $\) is the population standard deviation.

#### Given Data:

- Value of the data point (\( x \)): 75
- Population mean (\($ \mu $\)): 70
- Population standard deviation (\($ \sigma $\)): 5

#### Calculation:

\[$ z = \frac{75 - 70}{5} = \frac{5}{5} = 1 $\]

#### Interpretation:

The calculated z-score is \( $z = 1 $\). This indicates that the data point with a value of 75 is 1 standard deviation above the population mean of 70. In other words, the data point is relatively higher than the average value in the population by an amount equivalent to one standard deviation.

Z-scores are useful for comparing different datasets or understanding the relative position of individual data points within a distribution. A positive z-score signifies that the data point is above the mean, while a negative z-score would indicate a position below the mean.


### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

### Hypothesis Test: Effectiveness of a Weight Loss Drug

To determine if the weight loss drug is significantly effective, we will conduct a hypothesis test using a t-test.

#### Null Hypothesis (H₀):

The null hypothesis states that the mean weight loss due to the drug is zero, implying no effectiveness:
\[ $H₀: \mu = 0 $\]

#### Alternative Hypothesis (H₁):

The alternative hypothesis states that the drug is effective in causing weight loss:
\[ $H₁: \mu < 0$ \]

where \( \mu \) is the population mean weight loss.

#### Given Data:

- Sample size (\( n \)): 50
- Sample mean weight loss (\($ \bar{x} $\)): 6 pounds
- Sample standard deviation (\( s \)): 2.5 pounds

#### Confidence Level:

We will perform the hypothesis test at a 95% confidence level (\( \alpha = 0.05 \)).

#### Test Statistic (t-score):

The test statistic for a one-sample t-test is calculated as:
\[ t = \frac{\bar{x} - \mu₀}{\frac{s}{\sqrt{n}}} \]

where \( \mu₀ \) is the hypothesized population mean under the null hypothesis (typically 0 for effectiveness).

#### Calculation:

1. **Calculate the t-score:**
   \[ $t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} $\]
   \[$ t = \frac{6}{\frac{2.5}{\sqrt{50}}}$ \]
   \[$ t \approx \frac{6}{0.3536}$ \]
   \[$ t \approx 16.97$ \]

2. **Degrees of Freedom (df):**
   \($ df = n - 1 = 50 - 1 = 49 $\)


3. **Critical Value:**
   For a one-tailed t-test at \( $\alpha = 0.05$ \) and \($ df = 49 $\), the critical value from the t-distribution is approximately \($ t_{\alpha, df} = -1.676 $\).

#### Conclusion:

- **Compare t-score and Critical Value:**
  Since \( t = 16.97 \) is greater than \( -1.676 \), we reject the null hypothesis \( H₀ \).

- **Interpretation:**
  Based on the sample data, at a 95% confidence level, there is sufficient evidence to conclude that the weight loss drug is significantly effective. Participants in the study lost an average of 6 pounds, which is statistically significant given the standard deviation of 2.5 pounds.

Therefore, we reject the null hypothesis \( H₀ \) and accept the alternative hypothesis \( H₁ \), indicating that the weight loss drug is effective in causing weight loss among the participants in the study.


### Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

### Calculation of the 95% Confidence Interval for the True Proportion of People Satisfied with Their Job

To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can follow these steps:

1. **Identify the sample proportion (\($\hat{p}$\))**:
   \[$   \hat{p} = 0.65$   \]

2. **Identify the sample size (\(n\))**:
   \[$   n = 500$   \]

3. **Determine the standard error (SE) of the sample proportion**:
   \[
   SE =$ \sqrt{\frac{\hat{p} (1 - \hat{p})}{n}} $=$ \sqrt{\frac{0.65 \times (1 - 0.65)}{500}}$
   \]

4. **Find the z-score for a 95% confidence interval**:
   \[$z = 1.96$\]

5. **Calculate the margin of error (ME)**:
   \[$ ME $ = $ z \times SE $ = $ 1.96 \times \sqrt{\frac{0.65 \times 0.35}{500}}$\]

6. **Calculate the confidence interval (CI)**:
   \[$CI $= $\hat{p} \pm ME$ \]

### Calculations

#### Standard Error (SE)
\[$SE $= $\sqrt{\frac{0.65 \times 0.35}{500}} $= $\sqrt{\frac{0.2275}{500}} $= $\sqrt{0.000455} \approx 0.0213$
\]

#### Margin of Error (ME)
\[$ME $= $1.96 \times 0.0213 \approx 0.0418$\]

#### Confidence Interval (CI)
\[$CI $= $0.65 \pm 0.0418$\]

Therefore, the 95% confidence interval is:
\[$[0.65 - 0.0418, 0.65 + 0.0418]$ = $[0.6082, 0.6918]$\]

### Conclusion

The 95% confidence interval for the true proportion of people who are satisfied with their job is $[0.6082, 0.6918]$.


## Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

### Hypothesis Test for Teaching Methods

### Step 1: State the Hypotheses
- Null hypothesis ($H_0$): $\mu_A = \mu_B$ (The means of the two samples are equal, indicating no significant difference in performance.)
- Alternative hypothesis ($H_a$): $\mu_A \neq \mu_B$ (The means of the two samples are not equal, indicating a significant difference in performance.)

### Step 2: Gather the Sample Statistics
- Mean of Sample A ($\bar{X}_A$): 85
- Standard deviation of Sample A ($s_A$): 6
- Mean of Sample B ($\bar{X}_B$): 82
- Standard deviation of Sample B ($s_B$): 5
- Significance level ($\alpha$): 0.01

### Step 3: Calculate the Test Statistic
The test statistic for a two-sample t-test is given by:

\[$ t = \frac{\bar{X}_A - \bar{X}_B}{\sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}}$\]

### Step 4: Determine the Degrees of Freedom
The degrees of freedom for the t-test can be approximated using the following formula:

\[$ df = \frac{\left(\frac{s_A^2}{n} + \frac{s_B^2}{n}\right)^2}{\frac{\left(\frac{s_A^2}{n}\right)^2}{n-1} + \frac{\left(\frac{s_B^2}{n}\right)^2}{n-1}}$\]

### Step 5: Calculate the Critical Value and Compare
The critical value for a two-tailed test at $\alpha = 0.01$ can be found using a t-distribution table or a statistical calculator.

### Step 6: Make a Decision
Compare the calculated t-value with the critical t-value from the t-distribution table. If the absolute value of the calculated t-value is greater than the critical t-value, we reject the null hypothesis.

Let's perform these calculations using Python.



````python
import scipy.stats as stats
import math

# Given data
mean_A = 85
std_A = 6
mean_B = 82
std_B = 5
alpha = 0.01

# Assuming equal sample sizes for simplicity
n_A = n_B = 30  # Example sample size
n = n_A

# Calculate the t-statistic
t_stat = (mean_A - mean_B) / math.sqrt((std_A**2 / n) + (std_B**2 / n))

# Calculate the degrees of freedom
df = ( ((std_A**2 / n) + (std_B**2 / n))**2 ) / ( ((std_A**2 / n)**2 / (n - 1)) + ((std_B**2 / n)**2 / (n - 1)) )

# Determine the critical value from the t-distribution
critical_value = stats.t.ppf(1 - alpha/2, df)

# Print the results
t_stat, df, critical_value
````

Based on the calculated values:

t-statistic: $t_{\text{stat}}$
Degrees of Freedom: $df$
Critical Value: $t_{\text{critical}}$
If $|t_{\text{stat}}| > t_{\text{critical}}$, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

Ans: So, We fail to reject the null hypothesis, indicating that there is no significant difference in student performance between the two teaching methods at a significance level of 0.01.

Given:
- Population mean ($\mu$): 60
- Population standard deviation ($\sigma$): 8
- Sample mean ($\bar{X}$): 65
- Sample size ($n$): 50
- Confidence level: 90%

### Step 1: Calculate the Standard Error of the Mean (SEM)
The standard error of the mean is given by:

\[$\text{SEM} = \frac{\sigma}{\sqrt{n}}$\]

### Step 2: Determine the Z-Score for the Confidence Level
For a 90% confidence interval, the z-score can be found using a z-table or a statistical calculator. The z-score for a 90% confidence level is approximately 1.645.

### Step 3: Calculate the Margin of Error (ME)
The margin of error is given by:

\[$\text{ME} = Z \times \text{SEM}$\]

### Step 4: Determine the Confidence Interval
The confidence interval is calculated as:

\[$\text{CI} = \left( \bar{X} - \text{ME}, \bar{X} + \text{ME} \right)$\]

Let's perform these calculations using Python.


````python

import scipy.stats as stats
import math

# Given data
population_mean = 60
population_std = 8
sample_mean = 65
sample_size = 50
confidence_level = 0.90

# Step 1: Calculate the Standard Error of the Mean (SEM)
SEM = population_std / math.sqrt(sample_size)

# Step 2: Determine the Z-Score for the Confidence Level
z_score = stats.norm.ppf(0.5 + confidence_level / 2)

# Step 3: Calculate the Margin of Error (ME)
ME = z_score * SEM

# Step 4: Determine the Confidence Interval
confidence_interval = (sample_mean - ME, sample_mean + ME)

# Print the results
SEM, z_score, ME, confidence_interval
````

### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

## Hypothesis Test for Caffeine's Effect on Reaction Time

### Given:
- Sample size ($n$): 30
- Sample mean ($\bar{X}$): 0.25 seconds
- Sample standard deviation ($s$): 0.05 seconds
- Confidence level: 90%

### Step 1: State the Hypotheses
- Null hypothesis ($H_0$): $\mu = \mu_0$ (Caffeine does not have a significant effect on reaction time, where $\mu_0$ is the population mean reaction time without caffeine.)
- Alternative hypothesis ($H_a$): $\mu \neq \mu_0$ (Caffeine has a significant effect on reaction time.)

### Step 2: Calculate the Test Statistic
The test statistic for a one-sample t-test is given by:

\[$t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}}$\]

Since we are testing if caffeine has a significant effect, we will use the sample mean as the hypothesized population mean ($\mu_0$).

### Step 3: Determine the Degrees of Freedom
The degrees of freedom for the t-test is:

\[$df = n - 1$\]

### Step 4: Determine the Critical Value and Compare
The critical value for a two-tailed test at a 90% confidence level can be found using a t-distribution table or a statistical calculator.

### Step 5: Make a Decision
Compare the calculated t-value with the critical t-value from the t-distribution table. If the absolute value of the calculated t-value is greater than the critical t-value, we reject the null hypothesis.

Let's perform these calculations using Python.

````python

import scipy.stats as stats
import math

# Given data
sample_mean = 0.25
sample_std = 0.05
sample_size = 30
confidence_level = 0.90
population_mean = 0.0  # Assuming null hypothesis mean is 0 for no effect

# Step 1: Calculate the Standard Error of the Mean (SEM)
SEM = sample_std / math.sqrt(sample_size)

# Step 2: Calculate the t-statistic
t_stat = (sample_mean - population_mean) / SEM

# Step 3: Determine the Degrees of Freedom
df = sample_size - 1

# Step 4: Determine the Critical Value from the t-distribution
critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

# Print the results
t_stat, df, critical_value
````

Interpretation
Based on the calculations:

t-statistic: $t_{\text{stat}}$

Degrees of Freedom: $df$

Critical Value: $t_{\text{critical}}$

If $|t_{\text{stat}}| > t_{\text{critical}}$, we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

This will tell us if caffeine has a significant effect on reaction time at a 90% confidence level.

Ans:We reject the null hypothesis, indicating that caffeine has a significant effect on reaction time at a 90% confidence level.
